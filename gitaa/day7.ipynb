```python
# sets
it_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}
age = [22, 19, 24, 25, 26, 24, 25, 24]
```

```python
'''
Find the length of the set it_companies
'''
print(len(it_companies))
```

```python
'''
Add 'Twitter' to it_companies
'''
it_companies.add('Twitter')
print(it_companies)
```

```python
'''
Insert multiple IT companies at once to the set it_companies
'''
it_companies.update({'Tesla', 'Infosys', 'TCS'})
print(it_companies)
```

```python
'''
Remove one of the companies from the set it_companies
'''
it_companies.remove('IBM')
print(it_companies)
```

```python
'''
What is the difference between remove and discard
'''

it_companies.remove('Wipro')

```

```python
'''
Join A and B
'''
print(A.union(B))
```

```python
'''
Find A intersection B
'''
print(A.intersection(B))
```

```python
'''
Is A subset of B
'''
print(A.issubset(B))
```

```python
'''
Are A and B disjoint sets
'''
print(A.isdisjoint(B))
```

```python
'''
Join A with B and B with A
'''
print(A.union(B))
print(B.union(A))
```

```python
'''
What is the symmetric difference between A and B
'''
print(A.symmetric_difference(B))
```

```python
'''
Delete the sets completely
'''
del A
del B
```

```python
'''
Convert the ages to a set and compare the length of the list and the set, which one is bigger?
'''
age_set = set(age)
print("List length:", len(age))
print("Set length:", len(age_set))
print("Set is smaller because it removes duplicates.")
```

```python
'''
Explain the difference between the following data types: string, list, tuple and set
'''
String: Immutable sequence of characters.
List: Ordered and mutable collection allowing duplicates.
Tuple: Ordered but immutable collection allowing duplicates.
Set: Unordered, mutable, no duplicates.
```

```python
'''
I am a teacher and I love to inspire and teach people.
How many unique words have been used in the sentence?
'''
sentence = "I am a teacher and I love to inspire and teach people."
unique_words = set(sentence.lower().replace('.', '').split())
print("Unique word count:", len(unique_words))
print("Unique words:", unique_words)
```
