```python
def add_all_nums(*args):
    if all(isinstance(i, (int, float)) for i in args):
        return sum(args)
    else:
        return "All arguments must be numbers"
print(add_all_nums(2, 4, 6))
print(add_all_nums(2, 'hi', 3))
```

```python
def convert_celsius_to_fahrenheit(c):
    return (c * 9/5) + 32
print(convert_celsius_to_fahrenheit(0))
print(convert_celsius_to_fahrenheit(37))
```

```python
def check_season(month):
    month = month.lower()
    if month in ['september', 'october', 'november']:
        return 'Autumn'
    elif month in ['december', 'january', 'february']:
        return 'Winter'
    elif month in ['march', 'april', 'may']:
        return 'Spring'
    elif month in ['june', 'july', 'august']:
        return 'Summer'
    else:
        return 'Invalid month'
print(check_season('March'))
print(check_season('December'))
```

```python
def calculate_slope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)
print(calculate_slope(2, 4, 6, 10))
```

```python
import cmath
def solve_quadratic_eqn(a, b, c):
    d = cmath.sqrt(b**2 - 4*a*c)
    x1 = (-b + d) / (2*a)
    x2 = (-b - d) / (2*a)
    return (x1, x2)
print(solve_quadratic_eqn(1, -3, 2))
```

```python
def print_list(lst):
    for item in lst:
        print(item)
print_list(["apple", "banana", "mango"])
```

```python
def reverse_list(arr):
    rev = []
    for i in range(len(arr)-1, -1, -1):
        rev.append(arr[i])
    return rev
print(reverse_list([1, 2, 3, 4, 5]))
print(reverse_list(["A", "B", "C"]))
```

```python
def capitalize_list_items(lst):
    return [item.capitalize() for item in lst]
print(capitalize_list_items(["apple", "banana", "mango"]))
```

```python
def add_item(lst, item):
    lst.append(item)
    return lst
food_staff = ['Potato', 'Tomato', 'Mango', 'Milk']
print(add_item(food_staff, 'Meat'))
numbers = [2, 3, 7, 9]
print(add_item(numbers, 5))
```

```python
def remove_item(lst, item):
    if item in lst:
        lst.remove(item)
    return lst
numbers = [2, 3, 7, 9]
print(remove_item(numbers, 3))
food_staff = ['Potato', 'Tomato', 'Mango', 'Milk']
print(remove_item(food_staff, 'Mango'))

```

```python
def sum_of_numbers(n):
    return sum(range(1, n+1))
print(sum_of_numbers(5))
print(sum_of_numbers(10))
```

```python
def sum_of_odds(n):
    return sum(i for i in range(1, n+1) if i % 2 != 0)
print(sum_of_odds(10))
```

```python
def evens(n):
    evens = sum([i for i in range(n+1) if i % 2 == 0])
    print(evens)
    return f" The number of evens are {evens}."
print(evens(10))
```

```python
def evens_and_odds(n):
    evens = len([i for i in range(n+1) if i % 2 == 0])
    odds = n + 1 - evens
    return f"The number of odds are {odds}. The number of evens are {evens}."
print(evens_and_odds(100))
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact
print(factorial(5))
```

```python
def is_empty(param):
    return not bool(param)
print(is_empty([]))
print(is_empty("hi"))
```

```python
from statistics import mean, median, mode, variance, stdev
def calculate_mean(lst): return mean(lst)
def calculate_median(lst): return median(lst)
def calculate_mode(lst): return mode(lst)
def calculate_range(lst): return max(lst) - min(lst)
def calculate_variance(lst): return variance(lst)
def calculate_std(lst): return stdev(lst)
data = [4, 5, 6, 6, 7, 8, 9]
print(calculate_mean(data))
print(calculate_median(data))
print(calculate_mode(data))
print(calculate_range(data))
print(calculate_variance(data))
print(calculate_std(data))
```

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(is_prime(7))
print(is_prime(10))
```

```python
def all_unique(lst):
    return len(lst) == len(set(lst))
print(all_unique([1, 2, 3, 4]))
print(all_unique([1, 2, 2, 3]))
```

```python
def same_data_type(lst):
    return all(isinstance(i, type(lst[0])) for i in lst)
print(same_data_type([1, 2, 3]))
print(same_data_type([1, "hi", 3]))
```

```python
import keyword
def is_valid_variable(var):
    return var.isidentifier() and not keyword.iskeyword(var)
print(is_valid_variable("my_var"))
print(is_valid_variable("for"))
```

```python
from collections import Counter
from countries_dataset import countries_data

def most_spoken_languages(n=10):
    langs = [lang for country in countries_data for lang in country["languages"]]
    return Counter(langs).most_common(n)

def most_populated_countries(n=10):
    sorted_countries = sorted(countries_data, key=lambda x: x["population"], reverse=True)
    return sorted_countries[:n]

print("Top 10 Most Spoken Languages in the World:")
for lang, count in most_spoken_languages(10):
    print(f"{lang}: {count}")

print("\nTop 10 Most Populated Countries:")
for country in most_populated_countries(10):
    print(f"{country['name']}: {country['population']}")
```
