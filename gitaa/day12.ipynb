```python
import random
import string
```

```python
def random_user_id():
    return ''.join(random.choices(string.ascii_letters + string.digits, k=6))
print(random_user_id())
```

```python
def user_id_gen_by_user():
    char_count = int(input("Enter number of characters per ID: "))
    id_count = int(input("Enter number of IDs to generate: "))
    for _ in range(id_count):
        print(''.join(random.choices(string.ascii_letters + string.digits, k=char_count)))
user_id_gen_by_user()
```

```python
def rgb_color_gen():
    return f"rgb({random.randint(0,255)},{random.randint(0,255)},{random.randint(0,255)})"
print(rgb_color_gen())
```

```python
def list_of_hexa_colors(n):
    return ['#' + ''.join(random.choices('0123456789abcdef', k=6)) for _ in range(n)]
print(list_of_hexa_colors(5))
```

```python
def list_of_rgb_colors(n):
    return [f"rgb({random.randint(0,255)},{random.randint(0,255)},{random.randint(0,255)})" for _ in range(n)]
print(list_of_rgb_colors(5))
```

```python
def generate_colors(color_type, n):
    if color_type == 'hexa':
        return ['#' + ''.join(random.choices('0123456789abcdef', k=6)) for _ in range(n)]
    elif color_type == 'rgb':
        return [f"rgb({random.randint(0,255)},{random.randint(0,255)},{random.randint(0,255)})" for _ in range(n)]
print(generate_colors('hexa', 3))
print(generate_colors('hexa', 1))
print(generate_colors('rgb', 3))
print(generate_colors('rgb', 1))
```

```python
def shuffle_list(lst):
    random.shuffle(lst)
    return lst
print(shuffle_list([1, 2, 3, 4, 5, 6, 7]))
```

```python
def unique_random_numbers():
    return random.sample(range(10), 7)
print(unique_random_numbers())
```
