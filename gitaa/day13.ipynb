```python
numbers = [-4, -3, -2, -1, 0, 2, 4, 6]
neg_zero = [x for x in numbers if x <= 0]
print(neg_zero)
```

```python
list_of_lists = [[[1, 2, 3]], [[4, 5, 6]], [[7, 8, 9]]]
flattened = [num for sublist1 in list_of_lists for sublist2 in sublist1 for num in sublist2]
print(flattened)
```

```python
table = [(i, 1, i, i**2, i**3, i**4, i**5) for i in range(11)]
print(table)
```

```python
countries = [[('Finland', 'Helsinki')], [('Sweden', 'Stockholm')], [('Norway', 'Oslo')]]
flat_countries = [[country.upper(), country[:3].upper(), city.upper()] for [(country, city)] in countries]
print(flat_countries)
```

```python
countries = [[('Finland', 'Helsinki')], [('Sweden', 'Stockholm')], [('Norway', 'Oslo')]]
dict_countries = [{'country': country.upper(), 'city': city.upper()} for [(country, city)] in countries]
print(dict_countries)
```

```python
slope = lambda x1, y1, x2, y2: (y2 - y1) / (x2 - x1)
y_intercept = lambda x1, y1, x2, y2: y1 - slope(x1, y1, x2, y2) * x1

print("Slope:", slope(2, 3, 5, 11))
print("Y-intercept:", y_intercept(2, 3, 5, 11))
```
