```python
'''
Question 6:
Get length and width of a rectangle using prompt.
Calculate its area and perimeter.
'''
length = float(input("Enter length: "))
width = float(input("Enter width: "))
area = length * width
perimeter = 2 * (length + width)
print("Area:", area)
print("Perimeter:", perimeter)
```

```python
'''
Question 7:
Get radius of a circle using prompt.
Calculate area and circumference.
'''
r = float(input("Enter radius: "))
pi = 3.14
area = pi * r * r
circumference = 2 * pi * r
print("Area:", area)
print("Circumference:", circumference)
```

```python
'''
Question 8:
Calculate slope, x-intercept, and y-intercept of y = 2x - 2
'''
m = 2
y_intercept = -2
x_intercept = -y_intercept / m
print("Slope:", m)
print("Y-intercept:", y_intercept)
print("X-intercept:", x_intercept)
```

```python
'''
Question 9:
Find slope and Euclidean distance between points (2,2) and (6,10)
'''
x1, y1 = 2, 2
x2, y2 = 6, 10
m2 = (y2 - y1) / (x2 - x1)
distance = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
print("Slope:", m2)
print("Distance:", distance)
```
