```python
'''
The radius of a circle is 30 meters.
Calculate the area and circumference of the circle.(12)
'''
r = 30
pi = 3.15
area_of_circle = pi * r ** 2
circum_of_circle = 2 * pi * r
print(area_of_circle, circum_of_circle)
```

```python
'''
Take radius as user input and calculate the area.(12)
'''
r = float(input("Enter radius: "))
area = 3.14159 * r ** 2
print(area)
```
