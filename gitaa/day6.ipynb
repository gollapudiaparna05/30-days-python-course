```python
'''
Create an empty tuple
'''
empty_tuple = ()
print(empty_tuple)
```

```python
'''
Create a tuple containing names of your sisters and your brothers
'''
sisters = ('Anu', 'Harika')
brothers = ('Kiran', 'Raj')
print(sisters)
print(brothers)
```

```python
'''
Join brothers and sisters tuples and assign it to siblings
'''
siblings = sisters + brothers
print(siblings)
```

```python
'''
How many siblings do you have?
'''
print(len(siblings))
```

```python
'''
Modify the siblings tuple and add the name of your father and mother and assign it to family_members
'''
family_members = siblings + ('Father', 'Mother')
print(family_members)
```

```python
'''
Unpack siblings and parents from family_members
'''
sister1, sister2, brother1, brother2, father, mother = family_members
print(sister1)
print(sister2)
print(brother1)
print(brother2)
print(father)
print(mother)
```

```python
'''
Create fruits, vegetables and animal products tuples. Join the three tuples and assign it to a variable called food_stuff_tp.
'''
fruits = ('Apple', 'Banana', 'Orange')
vegetables = ('Carrot', 'Potato', 'Spinach')
animal_products = ('Milk', 'Eggs', 'Cheese')
food_stuff_tp = fruits + vegetables + animal_products
print(food_stuff_tp)
```

```python
'''
Change the food_stuff_tp tuple to a food_stuff_lt list
'''
food_stuff_lt = list(food_stuff_tp)
print(food_stuff_lt)
```

```python
'''
Slice out the middle item or items from the food_stuff_tp tuple or food_stuff_lt list
'''
mid = len(food_stuff_lt) // 2
if len(food_stuff_lt) % 2 == 0:
    print(food_stuff_lt[mid-1:mid+1])
else:
    print(food_stuff_lt[mid])
```

```python
'''
Slice out the first three items and the last three items from food_stuff_lt list
'''
print(food_stuff_lt[:3])
print(food_stuff_lt[-3:])
```

```python
'''
Delete the food_stuff_tp tuple completely
'''
del food_stuff_tp
```
