```python
'''
Create an empty dictionary called dog
'''
dog = {}
print(dog)
```

```python
'''
Add name, color, breed, legs, age to the dog dictionary
'''
dog['name'] = 'Buddy'
dog['color'] = 'Brown'
dog['breed'] = 'Labrador'
dog['legs'] = 4
dog['age'] = 3
print(dog)
```

```python
'''
Create a student dictionary and add
first_name, last_name, gender, age, marital status, skills,
country, city and address as keys for the dictionary
'''
student = {
    'first_name': 'Aparna',
    'last_name': 'Gollapudi',
    'gender': 'Female',
    'age': 20,
    'marital_status': 'Single',
    'skills': ['Python', 'Machine Learning'],
    'country': 'India',
    'city': 'Vizag',
    'address': 'Near kommadhi'
}
print(student)
```

```python
'''
Get the length of the student dictionary
'''
print(len(student))
```

```python
'''
Get the value of skills and check the data type, it should be a list
'''
skills_value = student['skills']
print(skills_value)
print(type(skills_value))
```

```python
'''
Modify the skills values by adding one or two skills
'''
student['skills'].extend(['HTML', 'CSS'])
print(student['skills'])
```

```python
'''
Get the dictionary keys as a list
'''
print(list(student.keys()))
```

```python
'''
Change the dictionary to a list of tuples using items() method
'''
print(list(student.items()))
```

```python
'''
Delete one of the items in the dictionary
'''
del student['marital_status']
print(student)
```
