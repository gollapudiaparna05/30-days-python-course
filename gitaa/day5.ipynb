```python
'''
Join the following lists:
front_end = ['HTML', 'CSS', 'JS', 'React', 'Redux']
back_end = ['Node','Express', 'MongoDB']
'''
front_end = ['HTML', 'CSS', 'JS', 'React', 'Redux']
back_end = ['Node', 'Express', 'MongoDB']
joined_list = front_end + back_end
print(joined_list)
```

```python
'''
After joining the lists, copy the joined list and assign it to full_stack,
then insert Python and SQL after Redux.
'''
full_stack = joined_list.copy()
index = full_stack.index('Redux') + 1
full_stack.insert(index, 'Python')
full_stack.insert(index + 1, 'SQL')
print(full_stack)
```

```python
'''
The following is a list of 10 students ages:
Sort the list and find the min and max age
'''
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
ages.sort()
print(ages)
print(min(ages))
print(max(ages))
```

```python
'''
Add the min age and the max age again to the list
'''
ages.append(min(ages))
ages.append(max(ages))
print(ages)
```

```python
'''
Find the median age
'''
ages.sort()
n = len(ages)
if n % 2 == 0:
    median = (ages[n//2 - 1] + ages[n//2]) / 2
else:
    median = ages[n//2]
print(median)
```

```python
'''
Find the average age
'''
average = sum(ages) / len(ages)
print(average)
```

```python
'''
Find the range of the ages
'''
age_range = max(ages) - min(ages)
print(age_range)
```

```python
'''
Compare (min - average) and (max - average)
'''
min_diff = abs(min(ages) - average)
max_diff = abs(max(ages) - average)
print(min_diff == max_diff)
```
